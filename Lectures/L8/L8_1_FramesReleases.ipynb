## Part 1 — Concepts and Notation

### What is a member release?

- A **member release** removes a force/moment transfer capability at an end of a member.
- Here we focus on **moment releases** (hinges).
- If an end is hinged, the **end moment must be zero** at that end.

Key constraint (local):
$$
M_b = 0 \quad \text{or} \quad M_e = 0
$$

### Member release types (classification)

**Figure placeholder:** Fig. 7.1 — Member release types

- **MT = 0:** no hinges (rigid at both ends, normal frame element)
- **MT = 1:** hinge at **beginning** (end **b**)
- **MT = 2:** hinge at **end** (end **e**)
- **MT = 3:** hinges at **both ends**

### Local DOFs and end forces (2D frame member)

Local displacement vector (member coordinates):

$$
\mathbf{u}^{T}
=
\begin{bmatrix}
u_1 & u_2 & u_3 & u_4 & u_5 & u_6
\end{bmatrix}
=
\begin{bmatrix}
u_b & v_b & \theta_b & u_e & v_e & \theta_e
\end{bmatrix}
$$

Local end-force vector:

$$
\mathbf{Q}^{T}
=
\begin{bmatrix}
Q_1 & Q_2 & Q_3 & Q_4 & Q_5 & Q_6
\end{bmatrix}
=
\begin{bmatrix}
N_b & V_b & M_b & N_e & V_e & M_e
\end{bmatrix}
$$

### Fixed-end force notation (FEFs)

We write local fixed-end forces using the convention:

- $Q_{xb}^F$ = FEF at node **b** in local **x** (axial) → corresponds to **DOF 1**
- $Q_{yb}^F$ = FEF at node **b** in local **y** (shear) → corresponds to **DOF 2**
- $Q_{mb}^F$ = FEF at node **b** (moment) → corresponds to **DOF 3**

- $Q_{xe}^F$ = FEF at node **e** in local **x** (axial) → corresponds to **DOF 4**
- $Q_{ye}^F$ = FEF at node **e** in local **y** (shear) → corresponds to **DOF 5**
- $Q_{me}^F$ = FEF at node **e** (moment) → corresponds to **DOF 6**

So the fixed-end force vector is
$$
\mathbf{Q}_f^{T}
=
\begin{bmatrix}
Q_{xb}^F & Q_{yb}^F & Q_{mb}^F &
Q_{xe}^F & Q_{ye}^F & Q_{me}^F
\end{bmatrix}
$$

## Part 2 — Derivations for Released Members

### Starting point (MT = 0): standard local stiffness relation

$$
\mathbf{Q} = \mathbf{k}\,\mathbf{u} + \mathbf{Q}^F
$$

Member releases modify **only**:
- the local stiffness matrix $\mathbf{k}$, and
- the fixed-end force vector $\mathbf{Q}^F$,

so the DSM workflow stays the same.

### Base local stiffness relations (no hinges)

Recall from last lecture. For a plane frame member with no hinges, the local end-force components can be written (explicitly) as:

$$
\begin{array}{rcl}
Q_1 & = & \dfrac{EA}{L}(u_1-u_4) + Q_{xb}^F \\
Q_2 & = & \dfrac{EI}{L^3}(12u_2+6Lu_3-12u_5+6Lu_6) + Q_{yb}^F \\
Q_3 & = & \dfrac{EI}{L^3}(6Lu_2+4L^2u_3-6Lu_5+2L^2u_6) + Q_{mb}^F \\

Q_5 & = & \dfrac{EI}{L^3}(-12u_2-6Lu_3+12u_5-6Lu_6) + Q_{ye}^F \\
Q_6 & = & \dfrac{EI}{L^3}(6Lu_2+2L^2u_3-6Lu_5+4L^2u_6) + Q_{me}^F
\end{array}
$$

$$
\mathbf{k}_{\text{frame}}
=
\begin{bmatrix}

### MT = 1 (hinge at b): enforce $M_b=0$

Hinge at the beginning means
$$
M_b = Q_3 = 0
$$
Start from
$$
Q_3
= \frac{EI}{L^3}(6Lu_2+4L^2u_3-6Lu_5+2L^2u_6) + Q_{mb}^F
$$
Set $Q_3=0$ and solve for $u_3$ (released rotation at b).

### MT = 1: solve for the released rotation $u_3$

Set $Q_3=0$:
$$
0= \frac{EI}{L^3}(6Lu_2+4L^2u_3-6Lu_5+2L^2u_6) + Q_{mb}^F
$$

Multiply by $\frac{L^3}{EI}$ and isolate $u_3$:
$$
4L^2u_3 = -6L(u_2-u_5) -2L^2u_6 - \frac{L^3}{EI}Q_{mb}^F
$$
$$
u_3 = -\frac{3}{2L}(u_2-u_5) - \frac{1}{2}u_6 - \frac{L}{4EI}Q_{mb}^F
$$


> **Key point**: $u_3$ is no longer an independent DOF in the member relations. The row/column corresponding to this DOF can be expressed in terms of a linear combination of other rows/columns in the matrix.

### MT = 1: substitute $u_3$ back into $Q_2, Q_5, Q_6$

Resulting modified equations (MT = 1):

$$
\begin{array}{rcl}
Q_1 & = & \dfrac{EA}{L}(u_1-u_4) + Q_{xb}^F \\
Q_2 & = & \dfrac{EI}{L^3}\left(3u_2-3u_5+3Lu_6\right)
          + \left(Q_{yb}^F - \dfrac{3}{2L}Q_{mb}^F\right) \\
Q_3 & = & 0 \\
Q_4 & = & \dfrac{EA}{L}(-u_1+u_4) + Q_{xe}^F \\
Q_5 & = & \dfrac{EI}{L^3}\left(-3u_2+3u_5-3Lu_6\right)
          + \left(Q_{ye}^F + \dfrac{3}{2L}Q_{mb}^F\right) \\
Q_6 & = & \dfrac{EI}{L^3}\left(3Lu_2-3Lu_5+3L^2u_6\right)
          + \left(Q_{me}^F - \dfrac{1}{2}Q_{mb}^F\right)
\end{array}
$$

row/column 3 are 0's:

$$
\mathbf{k}_{(MT=1)}
= \frac{EI}{L^3}
\begin{bmatrix}
\dfrac{AL^2}{I} & 0 & \boxed{0} & -\dfrac{AL^2}{I} & 0 & 0 \\
0 & 3 & \boxed{0} & 0 & -3 & 3L \\
\boxed{0} & \boxed{0} & \boxed{0} & \boxed{0} & \boxed{0} & \boxed{0} \\
-\dfrac{AL^2}{I} & 0 & \boxed{0} & \dfrac{AL^2}{I} & 0 & 0 \\
0 & -3 & \boxed{0} & 0 & 3 & -3L \\
0 & 3L & \boxed{0} & 0 & -3L & 3L^2
\end{bmatrix}
$$

row/column 3 are 0's:

$$
\mathbf{Q}^F_{(MT=1)}
=

F_{yb}^F - \frac{3}{2L}F_{mb}^F \\

F_{me}^F - \frac{1}{2}F_{mb}^F

Hinge at the end means
$$
M_e = Q_6 = 0
$$
Start from
$$
Q_6
= \frac{EI}{L^3}(6Lu_2+2L^2u_3-6Lu_5+4L^2u_6) + Q_{me}^F

Set $Q_6=0$ and solve for $u_6$ (released rotation at e).

### MT = 2: solve for the released rotation $u_6$

Set $Q_6=0$:
$$
0= \frac{EI}{L^3}(6Lu_2+2L^2u_3-6Lu_5+4L^2u_6) + Q_{me}^F
$$

Multiply by $\frac{L^3}{EI}$ and isolate $u_6$:
$$
4L^2u_6 = -6L(u_2-u_5) -2L^2u_3 - \frac{L^3}{EI}Q_{me}^F
$$
$$
u_6 = -\frac{3}{2L}(u_2-u_5) - \frac{1}{2}u_3 - \frac{L}{4EI}Q_{me}^F
$$

### MT = 2: substitute $u_6$ back into $Q_2, Q_3, Q_5$

Resulting modified equations (MT = 2):

$$
\begin{array}{rcl}
Q_1 & = & \dfrac{EA}{L}(u_1-u_4) + Q_{xb}^F \\[6pt]

Q_2 & = & \dfrac{EI}{L^3}\left(3u_2+3Lu_3-3u_5\right)
          + \left(Q_{yb}^F - \dfrac{3}{2L}Q_{me}^F\right) \\[6pt]

Q_3 & = & \dfrac{EI}{L^3}\left(3Lu_2+3L^2u_3-3Lu_5\right)
          + \left(Q_{mb}^F - \dfrac{1}{2}Q_{me}^F\right) \\[6pt]

Q_4 & = & \dfrac{EA}{L}(-u_1+u_4) + Q_{xe}^F \\[6pt]

Q_5 & = & \dfrac{EI}{L^3}\left(-3u_2-3Lu_3+3u_5\right)
          + \left(Q_{ye}^F + \dfrac{3}{2L}Q_{me}^F\right) \\[6pt]

Q_6 & = & 0
\end{array}

Row and column 6 are zero:

$$
\mathbf{k}_{(MT=2)}
=
\frac{EI}{L^3}
\begin{bmatrix}
\dfrac{AL^2}{I} & 0 & 0 & -\dfrac{AL^2}{I} & 0 & \boxed{0} \\
0 & 3 & 3L & 0 & -3 & \boxed{0} \\

\boxed{0} & \boxed{0} & \boxed{0} & \boxed{0} & \boxed{0} & \boxed{0}
\end{bmatrix}

Row/column 6 are zero:

$$
\mathbf{Q}_{f,(MT=2)}
=
\begin{bmatrix}
Q_{xb}^F \\
Q_{yb}^F - \dfrac{3}{2L}Q_{me}^F \\
Q_{mb}^F - \dfrac{1}{2}Q_{me}^F \\
Q_{xe}^F \\
Q_{ye}^F + \dfrac{3}{2L}Q_{me}^F \\
\boxed{0}
\end{bmatrix}

### MT = 3 (hinges at both ends): enforce $M_b=0$ and $M_e=0$

Hinges at both ends mean
$$
Q_3 = 0, \qquad Q_6 = 0
$$

We now have **two constraints** that couple $u_3$ and $u_6$.

Start from the base moment equations:
$$
Q_3= \frac{EI}{L^3}(6Lu_2+4L^2u_3-6Lu_5+2L^2u_6) + Q_{mb}^F
$$
$$
Q_6= \frac{EI}{L^3}(6Lu_2+2L^2u_3-6Lu_5+4L^2u_6) + Q_{me}^F
$$

### MT = 3: solve the coupled system for $u_3$ and $u_6$

Set $Q_3=0$ and $Q_6=0$, multiply both by $\frac{L^3}{EI}$:

$$
0 = 6Lu_2+4L^2u_3-6Lu_5+2L^2u_6 + \frac{L^3}{EI}Q_{mb}^F
$$
$$
0 = 6Lu_2+2L^2u_3-6Lu_5+4L^2u_6 + \frac{L^3}{EI}Q_{me}^F
$$

Solving gives (one convenient form):
$$
u_3 = -\frac{1}{L}(u_2-u_5) - \frac{L}{6EI}(2Q_{mb}^F - Q_{me}^F)
$$
$$
u_6 = -\frac{1}{L}(u_2-u_5) - \frac{L}{6EI}(2Q_{me}^F - Q_{mb}^F)
$$

### MT = 3: substitute $u_3$ and $u_6$ back $Q_2, Q_5$

$$
\begin{array}{rcl}
Q_1 & = & \dfrac{EA}{L}(u_1-u_4) + Q_{xb}^F \\[6pt]
Q_2 & = & Q_{yb}^F - \dfrac{1}{L}(Q_{mb}^F + Q_{me}^F) \\[6pt]
Q_3 & = & 0 \\[6pt]
Q_4 & = & \dfrac{EA}{L}(-u_1+u_4) + Q_{xe}^F \\[6pt]
Q_5 & = & Q_{ye}^F + \dfrac{1}{L}(Q_{mb}^F + Q_{me}^F) \\[6pt]
Q_6 & = & 0
\end{array}
$$

### Interpretation

- Notice that **$Q_2$ and $Q_5$ no longer contain any displacement terms**.
- They are **purely functions of the fixed-end forces**.
- The bending stiffness contributions have completely disappeared.
- This means the element **cannot develop bending resistance**.
- The transverse shear forces are determined solely by equilibrium of the fixed-end moments.
- The element now behaves as a **pure axial (truss-type) member**.

### MT = 3: stiffness matrix

Not only are row/column 3 and 6 are zero, but also row/column 2 and 5:

$$
\mathbf{k}_{(MT=3)}
=
\frac{EA}{L}
\begin{bmatrix}
1 & 0 & \boxed{0} & -1 & 0 & \boxed{0} \\
0 & 0 & \boxed{0} & 0 & 0 & \boxed{0} \\
\boxed{0} & \boxed{0} & \boxed{0} & \boxed{0} & \boxed{0} & \boxed{0} \\
-1 & 0 & \boxed{0} & 1 & 0 & \boxed{0} \\
0 & 0 & \boxed{0} & 0 & 0 & \boxed{0} \\
\boxed{0} & \boxed{0} & \boxed{0} & \boxed{0} & \boxed{0} & \boxed{0}
\end{bmatrix}
$$

What does this stiffness matrix remind you of?

### MT = 3: fixed-end force vector

Row/column 3 and 6 are zero:

$$
\mathbf{Q}^F_{(MT=3)}
=
\begin{bmatrix}
Q_{xb}^F \\
Q_{yb}^F - \frac{1}{L}(Q_{mb}^F + Q_{me}^F) \\
\boxed{0} \\
Q_{xe}^F \\
Q_{ye}^F + \frac{1}{L}(F_{mb}^F + Q_{me}^F) \\
\boxed{0}
\end{bmatrix}
$$

### Implementation note: keep 6×6, but released DOFs have zero rows/cols

- Even though releases reduce the **independent** member coordinates, we often keep a **6×6** member form.
- Released moment rows/cols (and their corresponding fixed-end terms) become **zero** in the stiffness matrix and FEF vector.
- This keeps the DSM assembly code clean: same DOF mapping size for every member.

Then transform each element to global coordinates as usual:

$$
\mathbf{K} = \mathbf{T}^T \mathbf{k} \mathbf{T},
\qquad
\mathbf{F}^F = \mathbf{T}^T \mathbf{Q}^F
$$

## Part 3 — Hinged Joints and Matrix Singularity

### Hinged joints in beams and plane frames

A **hinged joint** is a joint where all members meeting at that joint are connected by **moment releases**.

<!-- TODO: Insert textbook Fig. 7.2 (hinged joint example) -->

**Figure placeholder:** Fig. 7.2 — Example plane frame showing a hinged joint

Physical consequence:

- The joint **cannot transmit bending moment**.
- The joint is **free to rotate**.
- Axial and shear forces may still be transmitted.


### What does this mean for the DOFs?

At a hinged joint:

- The rotational DOF $\theta_j$ is not constrained (it is free).
- But no connected member provides **rotational stiffness**.
- The joint rotation does not generate bending moment.

In other words: The structure has a rotation that costs **zero strain energy**.


### DSM partitioned equilibrium

Recall the partitioned system:

$$
\begin{bmatrix}
\mathbf{K}_{ff} & \mathbf{K}_{fr}\\
\mathbf{K}_{rf} & \mathbf{K}_{rr}
\end{bmatrix}
\begin{bmatrix}
\mathbf{u}_f\\
\mathbf{u}_r
\end{bmatrix}
=
\begin{bmatrix}
\mathbf{F}_f\\
\mathbf{F}_r
\end{bmatrix}
$$

We solve for the unknown displacements from

$$
\mathbf{u}_f
=
K_{ff}^{-1} (\mathbf{F}_f - \mathbf{K}_{fr}\mathbf{u}_r)
$$

This requires that $\mathbf{K}_{ff}$ be **invertible**.


### What happens if $\theta_j$ is left in the free set?

The rotational DOF $\theta_j$ is not restrained, so it is placed in the free displacement vector $\mathbf{u}_f$.

As a free DOF, it has a corresponding row and column in the stiffness matrix $\mathbf{K}_{ff}$.

However, at a fully hinged joint, no connected element provides rotational stiffness.  
From our earlier derivations, the entire row and column associated with $\theta_j$ are zero.

Therefore:

- $\mathbf{K}_{ff}$ contains a **zero row and column**.
- $\mathbf{K}_{ff}$ becomes **singular** (non-invertible).
- The system of equations does not admit a unique displacement solution.

### Important clarification

This situation does **not** indicate a physical instability.

The structure may be completely stable in translation and capable of carrying load.

The singularity arises because a rotational DOF was left in the free set **without any associated stiffness**.

In other words:

- The joint can rotate freely.
- That rotation does not generate internal moment.
- The stiffness matrix has no resistance in that direction.

The issue is therefore **mathematical**, not structural.

To obtain a well-posed system of equations, this DOF must be treated properly in the DSM formulation.

### Two Modeling Approaches

Two practical ways to avoid singularity:

**Method A — Artificially restrain (fix) the joint rotation**
- Put the hinged-joint rotation into the **restrained set** $\mathbf{u}_r$ (not in $\mathbf{u}_f$)
- The computed reaction moment at that DOF ends up being **zero** anyway

**Method B — Artifically stiffen the joint DOF**
- Model the joint as rigidly connected to **one** of the members meeting there
- Assign the joint three DOFs $(u_x, u_y, \theta)$ as usual
- Other members connect by hinges (moment releases) so the net joint moment remains zero


## Part 4 - Method A, Fictitious Fixity 


###  The Main Idea

The most straightforward remedy for the singularity is:

> Eliminate the rotational DOF of hinged joints by modeling such joints as restrained (fixed) against rotation.

This is imposing **fictitious fixity** that does not change global structural behavior.

### Internal Moment at a Fully Hinged Joint

At a fully hinged joint:

- All connected members have **moment releases** at the node.
- The joint **cannot transmit bending moment**.
- Therefore, each connected element, $i$, contributes  
  $$
  M_i = 0 \quad \text{at the joint.}
  $$


Joint equilibrium must always satisfy

$$
\sum_{i=1}^{n_{ele}} M_i + M_{\text{ext}} = 0
$$

where:

- $M_i$ = internal end moment from element $i$
- $M_{\text{ext}}$ = applied or reaction moment at the joint

<!-- For a fully hinged joint:

- No external reaction moment can develop  
  ($M_{\text{ext}} = 0$)
- Each released member satisfies $M_i = 0$ -->

### Artificially Fixing the Rotation

Suppose we introduce **fictitious fixity** by restraining the rotation at the joint DOF associated with member $j$:

$$
\theta_j = 0
$$

Instead of leaving $\theta_j$ in the free displacement vector, we move it into the restrained set.

After solving for the free DOFs, reactions $F_r$ are computed at all restrained DOFs. For the fictitiously fixed rotation for member $j$, the reported reaction is a moment:

$$
M_{\text{ext}} = M_j
$$



### What Happens to the “Reaction” at That DOF?

As stated, joint equilibrium requires

$$
\sum_{i=1}^{n_{ele}} M_i + M_{\text{ext}} = 0
$$

Separating the contribution from member $j$ from the internal moment sum and setting $M_{\text{ext}} = M_j$.

$$
\left( \sum_{\substack{i=1 \\ i \ne j}}^{n_{ele}} M_i \right) + M_j = 0.
$$

At a hinged joint, all other connected members contribute zero moment, so $M_j = 0$

Thus, the reported “reaction” at the fictitiously fixed member, $j$, is a numerical artifact and evaluates to zero (up to roundoff).

### Why the Structural Response is Unchanged

Imposing $\theta_j = 0$ does not change the moments developed in the joint, because:

- No connected element provides rotational stiffness.
- The global stiffness matrix contains no coupling terms involving $\theta_j$.
- The equilibrium equations do not depend on that rotational DOF.

Because no bending stiffness exists at the joint, fixing the rotation does not generate any additional reaction moment (moment is related to displacement through stiffness).

You can think of DOF $j$ as a restrained DOF that simply develops **no reaction load**.

### Implementing with DSM Partitioning

Instead of leaving $\theta_j$ in the free displacement vector $\mathbf{u}_f$,

we classify it as a **restrained DOF**.

So:

- It is placed in $\mathbf{u}_r$.
- It is removed from $\mathbf{K}_{ff}$.
- It is no longer part of the system we invert:

$$
\mathbf{K}_{ff}^{-1}
$$

The zero-stiffness row and column disappear from the matrix being solved.

### Method A Summary

Method A is therefore a **numerical bookkeeping device**:

- It removes a zero-stiffness DOF from the unknown set,
- Eliminates the singularity in $\mathbf{K}_{ff}$,
- Preserves the correct internal force distribution.

The physics is unchanged but the algebra becomes well-posed.

### Method A — Step-by-Step Procedure

1. **Identify the hinged joint**

   - All connected member ends at that node are moment-released.
   - The joint cannot transmit bending moment.

2. **Classify the rotation as a restrained DOF**

   - Place $\theta_j$ in the restrained displacement vector $\mathbf{u}_r$.
   - Impose the artificial condition
     $$
     \theta_j = 0
     $$

   This removes the hinge rotation from the unknown set, $u_f$.



3. **Assemble the global stiffness matrix normally**

   - Because all connected elements are released,
     the assembled stiffness terms associated with $\theta_j$ are zero.
   - No rotational stiffness is introduced.


4. **Solve for the free DOFs**

   - $\theta_j$ is not included in $\mathbf{u}_f$.
   - The matrix $\mathbf{K}_{ff}$ no longer contains a zero row/column.
   - The system becomes invertible.

5. **Compute reactions**

   Reactions are obtained from

   $$
   \mathbf{F}_r
   =
   \mathbf{K}_{rf}\mathbf{u}_f
   +
   \mathbf{K}_{rr}\mathbf{u}_r
   +
   \mathbf{F}_r^{\,F}
   $$

   The reaction moment at the fictitiously fixed DOF evaluates to

   $$
   M_j \approx 0
   $$

   confirming hinge behavior.

## Part 5 - Method B, Fictitious Stiffness

### The Main Idea

Instead of removing the joint rotation from the free system (Method A), we:

> Model **one (and only one)** of the members meeting at that joint as having rotational rigidity.

All other members framing into the joint remain **hinged (moment-released)**.

**Figure placeholder:** Joint $j$ rigid to one member, hinged to others


### Why This Removes the Singularity

By making **one member rigidly connected**, that member contributes:

$$
\frac{4EI}{L}
$$

type rotational stiffness at the joint.

Therefore:

- The joint rotation DOF now has stiffness.
- $\mathbf{K}_{ff}$ is no longer singular.
- The system becomes solvable without fictitious restraints.

### Why This Does NOT Change the Structural Behavior

Even though one member, $j$, is fictitiously modeled as rigid at the joint:

- There is **no external moment applied at that joint**.
- All other members connected there are hinged.
- Hinged ends cannot transmit moment.

From joint equilibrium:

$$
\sum_{i=1}^{n_{ele}} M_i = 0
$$

All hinged members satisfy $M_i = 0$, so the joint moment equilibrium can be written as:

$$
\sum_{i=1}^{n_c} M_i
=
\left( \sum_{\substack{i=1 \\ i \ne j}}^{n_c} M_i \right)
+ M_j
=
0 + M_j
=
0.
$$

Thus, the rigidly connected member must also have zero end moment, $M_j = 0$, in the final solution in order to satisfy joint equilibrium.


### Key Mechanical Insight

Although we introduced rotational stiffness at the joint:

- The structure has no moment demand at that location.
- Static equilibrium forces the joint moment to zero.
- The internal force distribution remains unchanged.

We have:

- Stabilized the algebra (no singular matrix),
- Preserved correct statics,
- Avoided artificial restraints.

### What Method B Really Is

Method B is a **modeling device** that:

- Supplies stiffness to a mechanism DOF,
- Keeps the joint rotation as a free unknown,
- Relies on equilibrium to enforce zero moment at the joint (i.e., correct hinge behavior).

The structural response remains the same. The numerical system becomes well-posed.

### Method B — Step-by-Step Procedure

1. **Identify the fully hinged joint**

   - Confirm that all members meeting at the joint are moment-released.
   - Recognize that this creates a rotational DOF with zero stiffness.

2. **Select one “reference” member at the joint**

   - Choose **one (and only one)** connected member.
   - Model that member as **rigidly connected** at the joint (no moment release at that end).
   - The rotation DOF is now supported by this member’s rotational stiffness.

   This supplies stiffness to the otherwise unsupported rotation.

3. **Keep all other members hinged at that joint**

   - Use the appropriate released-end formulation (MT = 1 or MT = 2).
   - Their end moments at the joint are explicitly zero.

   These members do **not** contribute rotational stiffness at the joint.

4. **Treat the joint as a normal rigid node in global bookkeeping**

   - The rotational DOF $\theta_j$ remains in the **free displacement vector**.
   - Because one member now contributes stiffness,  
     the corresponding row/column in $\mathbf{K}_{ff}$ is no longer zero.

   The singularity disappears.

5. **Assemble and solve the system**

   - Assemble the global stiffness matrix in the usual way.
   - Solve for the free DOFs.
   - Compute member end forces.

   Although one member was modeled as rigid at the joint,  
   equilibrium at that joint still requires:

   $$
   \sum M = 0
   $$

   Since the other members are hinged and no external moment is applied,  
   the rigid member’s end moment evaluates to:

   $$
   M_j \approx 0
   $$

### Method A vs Method B — Clear Takeaway

<div style="display:flex; gap:3rem; align-items:flex-start;">

<div style="flex:1;">

**Method A — Fictitious Fixity**

- Reclassify the joint rotation DOF as **restrained**.

**Effect:**
- Eliminates the zero-stiffness DOF from $\mathbf{K}_{ff}$.
- The computed fictituous reaction moment evaluates to $M_j = 0$

</div>

<div style="flex:1;">

**Method B — Fictitious Stiffness**

- Keep the joint rotation DOF **free**.
- Make **one (and only one)** connected member rigid at the joint.
- All other connected members remain hinged.

**Effect:**
- The rigid member supplies nonzero rotational stiffness.
- From joint equilibrium,$\sum M = 0$, the fictituously rigid member’s end moment evaluates to zero.

</div>

</div>


### What Is Different?

- **Method A:** enforces $\theta_j = 0$
- **Method B:** allows $\theta_j \neq 0$
  (determined numerically from solving $u_f$)

Therefore, local element rotations can be slightly different across methods, but this does not affect the internal force distribution.

### Key Takeaway

Both methods represent the **same physical structure** and satisfy the same equilibrium and compatibility conditions.

- Internal force distribution → **the same**
- Physically meaningful deformations → **the same**
- Reported joint rotation at the hinges → **may differ**

The difference is numerical, not structural.

### Member 3 (MT = 1)

At the start of member 3 there is a **pin support**.

We have two equivalent modeling choices:

**Option 1 — Standard Frame Element (MT = 0)**
- Use the full $6 \times 6$ local stiffness matrix (no releases).
- Impose $u_x = 0$ and $u_y = 0$ in the restrained DOF set.
- Leave the rotational DOF $\theta$ free.

In this approach, this location remains numerically stable without any additional treatment.

**Option 2 — Member Release at Start (MT = 1)**
- Model the pin as a **hinge at the start of the element**.
- This removes moment transfer at that end.
- The element is therefore classified as **MT = 1**.

Because a rotational release eliminates bending stiffness at that end, this creates another situation where fictitous fixity must be applied



In this example, Option 2 is convenient because:

- Geometrically, member 3 matches member 2.  
- Same release condition (hinge at $b$).  
- So we can reuse:

$$
\mathbf{k}_{3} = \mathbf{k}_{2}
$$

- No member loading:

$$
\mathbf{Q}_{3}^F = \mathbf{0}
$$

- Vertical member:

$$
\theta = 90^\circ \quad \mathbf{T}_{3} = \mathbf{T}_{2}
$$

- Mapping
$$
[1,2,3,4,5,6]_{\text{local}} = [10,11,12,7,8,9]_{\text{global}}
$$


Note: Both approaches are correct — using MT = 1 simply keeps the setup cleaner.
