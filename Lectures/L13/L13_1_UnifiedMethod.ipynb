## Part 6 - Back to the Unified Framework

> connecting Schek to Veenendaal and Block

### where force density sits among other methods

In the Veenendaal and Block comparison, force density is one member of a family of form-finding procedures.

Very loosely:

- **force density** uses prescribed $q_i = s_i/l_i$ and solves directly for geometry
- **dynamic relaxation** introduces pseudo-dynamics and converges iteratively to equilibrium
- **stiffness method** uses element constitutive behavior and solves a nonlinear equilibrium problem

So these methods differ in:
- which variables are treated as primary
- whether the update is direct or iterative
- how the internal forces are related to the geometry

The force density method is especially attractive because it turns equilibrium into a **linear algebra problem** at the nodal level.

### relation to what we have done in this course

There is also a useful conceptual bridge to the direct stiffness method:

In DSM for trusses, equilibrium is typically written in terms of:
- nodal displacements
- element axial deformations
- element axial forces from constitutive behavior

In force density, the formulation is more geometric:
- unknowns are directly the node coordinates
- connectivity is encoded by $\\mathbf{C}$
- the "material-like" parameter is the prescribed force density $q_i$

So although the algebra looks different, both approaches are still enforcing **nodal equilibrium of an interconnected bar or cable network**.

That is exactly why the Veenendaal and Block paper is helpful:
it clarifies how methods that look very different are actually closely related.
