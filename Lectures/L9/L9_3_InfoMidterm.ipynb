Overview of logistics, expectations, and topics.

### Logistics

- **Date:** Friday, March 20
- **Time:** During normal lecture time
- **Length:** ~2 hours

The exam will focus primarily on applying the **Direct Stiffness Method (DSM)** to one of the structural systems covered in recent weeks.

### Exam Structure

The midterm will consist of **four parts**:

1. **Multiple choice** — 15%
2. **Short answer / conceptual questions** — 15%
3. **Long calculation problem** — 60%
4. **Code comprehension / interpretation** — 10%

The majority of the exam will be **Part 3**, which will walk through the implementation of the **Direct Stiffness Method** step-by-step for a structural system.

### Long Calculation Question

The main problem will involve applying the **DSM workflow**, broken into several smaller parts.

Typical tasks may include:

- calculating FEFs
- assembling the global stiffness matrix
- assembling the global force and FEF vectors
- applying boundary conditions
- solving for unknown displacements
- computing reactions and element forces
- applying support settlement

You should be comfortable with the **full DSM pipeline**, even if individual steps are simplified.
