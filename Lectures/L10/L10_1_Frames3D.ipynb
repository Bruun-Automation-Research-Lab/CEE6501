## Part 1 — DSM Element Roadmap

> From trusses to 2D beams to 2D frames to 3D frames

### The 3D Frame Element (Where We Are Headed)

- A general 3D frame member has **6 DOFs per node** (12 per element)
  - translations: $u_x, u_y, u_z$
  - rotations: $\theta_x, \theta_y, \theta_z$
- This single element can represent **axial**, **torsion**, and **bending** behavior

### Key Idea: Superposition

You can think of the 3D frame element as a **superposition** of four behaviors:

1. **Axial** deformation (truss-like)
2. **Torsion** about the member axis
3. **Bending** about one principal axis (Z-axis)
4. **Bending** about the other principal axis (Y-axis)

## Part 2 — Pure Torsion Member

> The missing 1D ingredient needed to build the full 3D frame element

### Key Analogy: Torsion Looks Just Like Axial

- Mathematically, the **pure torsion member** is directly analogous to the **axial bar element**
- The variable pair changes from
  - axial: force $F_x$ and displacement $u_x$
  - torsion: torque $M_x$ and rotation $\theta_x$
- So the same 2-node, 1-DOF-per-node structure appears again

<div style="text-align:center;">
  <img src="assets/L10_Axial_vs_Torsion.png" style="width:88%;">
</div>


### Same Mathematics, Different Physics

- In an **axial** member, the bar stretches or shortens
- In a **torsional** member, the shaft twists about its longitudinal axis
- The nodal DOFs for the torsion element are therefore the end rotations

$$
v =
\begin{Bmatrix}
\theta_{x1} \\
\theta_{x2}
\end{Bmatrix}
$$

- The corresponding nodal actions are the end torques

$$
q =
\begin{Bmatrix}
M_{x1} \\
M_{x2}
\end{Bmatrix}
$$


### Torsional Constitutive Relation

For a prismatic shaft under pure torsion,

$$
\beta = \frac{M_x}{GJ}
$$

where:

- $\beta$ = rate of twist
- $M_x$ = internal torque about the member axis
- $G$ = shear modulus (modulus of rigidity)
- $J$ = torsional constant of the cross section

> For a circular shaft, $J$ equals the polar moment of inertia. In general, it is the **torsional constant**, not necessarily the polar moment.

### End Rotation From Integrating the Twist

Assuming constant $GJ$ and constant internal torque over the element length,

$$
\theta_{x2} = \int_0^L \beta\,dx
= \int_0^L \frac{M_x}{GJ}\,dx
= \frac{M_x L}{GJ}
$$

So, by direct comparison with the axial bar,

$$
[d] = \frac{L}{GJ},
\qquad
[\Phi] = -1
$$

This is exactly the same mathematical structure as the axial element, with $EA$ replaced by $GJ$ and $u_x$ replaced by $\theta_x$.

### Torsional Element Stiffness Matrix

Therefore, the local torsional stiffness matrix is

$$
[k] = \frac{GJ}{L}
\begin{bmatrix}
1 & -1 \\
-1 & 1
\end{bmatrix}
$$

and the element force-displacement relation becomes

$$
\begin{Bmatrix}
M_{x1} \\
M_{x2}
\end{Bmatrix}
=
\frac{GJ}{L}
\begin{bmatrix}
1 & -1 \\
-1 & 1
\end{bmatrix}
\begin{Bmatrix}
\theta_{x1} \\
\theta_{x2}
\end{Bmatrix}
$$


### Takeaway for 3D Frame Assembly

- The torsion member is the **rotational analogue** of the axial bar
- It contributes stiffness only in the member-axis rotational DOFs
- In the full 3D frame element, this torsional block will sit alongside:
  - the axial block
  - the bending-about-$y$ block
  - the bending-about-$z$ block

So the full 12×12 local stiffness matrix is built by **superposing these familiar pieces**.
