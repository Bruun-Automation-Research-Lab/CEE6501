## Part 2 — Pure Torsion Member

> The missing 1D ingredient needed to build the full 3D frame element

### Key Analogy: Torsion Looks Just Like Axial

- Mathematically, the **pure torsion member** is directly analogous to the **axial bar element**
- The variable pair changes from
  - axial: force $q_x$ and displacement $v_x$
  - torsion: torque $q_{mx}$ and rotation $v_{\theta x}$
- So the same 2-node, 1-DOF-per-node structure appears again

<div style="text-align:center;">
  <img src="assets/L1_AxialTorsion.png" style="width:88%;">
</div>


### Same Mathematics, Different Physics

- In an **axial** member, the bar stretches or shortens
- In a **torsional** member, the shaft twists about its longitudinal axis
- The nodal DOFs for the torsion element are therefore the end rotations

$$
v =
\begin{Bmatrix}
\theta_{x1} \\
\theta_{x2}
\end{Bmatrix}
$$

- The corresponding nodal actions are the end torques

$$
q =
\begin{Bmatrix}
m_{x1} \\
m_{x2}
\end{Bmatrix}
$$


### Torsional Constitutive Relation

For a prismatic shaft under pure torsion,

$$
\beta = \frac{m_x}{GJ}
$$

where:

- $\beta$ = rate of twist
- $m_x$ = internal torque about the member axis
- $G$ = shear modulus (modulus of rigidity) = 
- $J$ = torsional constant of the cross section

$$
G = \frac{E}{2(1+\nu)}
$$

where $\nu$ is Poisson’s ratio.

### End Rotation From Integrating the Twist

Assuming constant $GJ$ and constant internal torque over the element length,

$$
\theta_{x2} = \int_0^L \beta\,dx
= \int_0^L \frac{m_x}{GJ}\,dx
= \frac{m_x L}{GJ}
$$


### Comparison With Axial Bar,

$$
\theta_x = m_x \frac{L}{GJ},
\qquad
v_x = q_x \frac{L}{EA},
$$

| element type | flexibility, $f$ | stiffness, $k$ |
|---|---:|---:|
| axial member | $\dfrac{L}{EA}$ | $\dfrac{EA}{L}$ |
| torsion member | $\dfrac{L}{GJ}$ | $\dfrac{GJ}{L}$ |

This is exactly the same mathematical structure as the axial element, with $EA$ replaced by $GJ$ and $v_x$ replaced by $\theta_x$.

### Torsional Element Stiffness Matrix

Therefore, the local torsional stiffness matrix is

$$
k = \frac{GJ}{L}
\begin{bmatrix}
1 & -1 \\
-1 & 1
\end{bmatrix}
$$

and the element force-displacement relation becomes

$$
\begin{Bmatrix}
m_{x1} \\
m_{x2}
\end{Bmatrix}
=
\frac{GJ}{L}
\begin{bmatrix}
1 & -1 \\
-1 & 1
\end{bmatrix}
\begin{Bmatrix}
\theta_{x1} \\
\theta_{x2}
\end{Bmatrix}
$$

## Part 3 — Complete 3D Frame Element

> Assembling axial, torsion, and two bending behaviors into one 12-DOF local element

### Deriving the Local Stiffness Matrix

In principle, we could derive the full local stiffness matrix using the **unit displacement method**, just as we have done for smaller elements.

However, for a **12×12 3D frame element**, that process becomes lengthy and not especially insightful.

Instead, we take advantage of the fact that many DOFs are **uncoupled by physical behavior**, and build the full local matrix by **superposing familiar component matrices**:

- axial deformation
- torsion
- bending about the local $y$-axis
- bending about the local $z$-axis


### Superposition Strategy

The complete local 3D frame element is obtained by placing the four component behaviors into the appropriate DOF locations within a single 12×12 matrix.

- the **axial block** contributes stiffness in the local $x$-translation DOFs
- the **torsional block** contributes stiffness in the local $x$-rotation DOFs
- the **bending-about-$y$ block** couples $v_z$ and $\theta_y$
- the **bending-about-$z$ block** couples $v_y$ and $\theta_z$

So rather than deriving one large matrix from scratch, we assemble it from these previously derived building blocks.

### Local Element DOF Numbering


We use the standard **local 3D frame element** ordering, with **6 DOFs at each node** for a total of **12 element DOFs**.

Here:

- **node b** = begin node  
- **node e** = end node

At each node, the local DOFs are ordered as:

- translations in the local $x$, $y$, and $z$ directions
- rotations about the local $x$, $y$, and $z$ axes

So the ordering is:

$$
(x,\; y,\; z,\; \theta x,\; \theta y,\; \theta z)
$$

for node $b$, followed by the same ordering for node $e$.



### Local Force Vector

$$
\begin{aligned}
\mathbf{q}^T
&=
\begin{bmatrix}
q_1 & q_2 & q_3 & q_4 & q_5 & q_6 & q_7 & q_8 & q_9 & q_{10} & q_{11} & q_{12}
\end{bmatrix}
\\[6pt]
&=
\begin{bmatrix}
q_{x,b} & q_{y,b} & q_{z,b} & q_{mx,b} & q_{my,b} & q_{mz,b} & q_{x,e} & q_{y,e} & q_{z,e} & q_{mx,e} & q_{my,e} & q_{mz,e}
\end{bmatrix}
\end{aligned}
$$

### Local Displacement Vector

$$
\begin{aligned}
\mathbf{v}^T
&=
\begin{bmatrix}
v_1 & v_2 & v_3 & v_4 & v_5 & v_6 & v_7 & v_8 & v_9 & v_{10} & v_{11} & v_{12}
\end{bmatrix}
\\[6pt]
&=
\begin{bmatrix}
v_{x,b} & v_{y,b} & v_{z,b} & v_{\theta x,b} & v_{\theta y,b} & v_{\theta z,b} & v_{x,e} & v_{y,e} & v_{z,e} & v_{\theta x,e} & v_{\theta y,e} & v_{\theta z,e}
\end{bmatrix}
\end{aligned}
$$

- the local $x$-axis is taken along the member centroidal axis
- the local $y$- and $z$-axes are aligned with the member cross-sectional principal directions
- the local axes are oriented to form a **right-handed** coordinate system

### Global Joint Degrees of Freedom

A free joint in a space frame can:

- translate in the global $X$, $Y$, and $Z$ directions
- rotate about the global $X$, $Y$, and $Z$ axes

So each unconstrained joint has **6 global degrees of freedom**.

$$
\text{DOFs per joint} = 6
$$

Therefore, for a structure with $N_J$ joints and $N_R$ restrained coordinates,

$$
\mathrm{NDOF} = 6N_J - N_R
$$

A stable space-frame model must include enough independent support restraints to eliminate the **6 rigid body modes** of a free 3D body: three translations and three rotations.

### 1-Based Global DOF Numbering

Global joint loads and displacements are defined with respect to a global right-handed $XYZ$ coordinate system.

Using **1-based numbering**, the 6 DOFs associated with node $n$ are:

$$
\begin{aligned}
6n - 5 &\;\rightarrow\; X \text{ translation} \\
6n - 4 &\;\rightarrow\; Y \text{ translation} \\
6n - 3 &\;\rightarrow\; Z \text{ translation} \\
6n - 2 &\;\rightarrow\; \text{rotation about } X \\
6n - 1 &\;\rightarrow\; \text{rotation about } Y \\
6n     &\;\rightarrow\; \text{rotation about } Z
\end{aligned}
$$

This gives a consistent way to build the global displacement vector, global force vector, and element connectivity mappings.

### Resolving Global Forces Into Local Components

The local force components at the start node, $b$, are obtained by projecting the global force vector onto the local element axes.

If the global force vector is $\mathbf{f}^T = [f_1,f_2,f_3]^T$

then the corresponding local components are

$$
\begin{aligned}
q_1 &= f_1 \cos \theta_{xX} + f_2 \cos \theta_{xY} + f_3 \cos \theta_{xZ} \\
q_2 &= f_1 \cos \theta_{yX} + f_2 \cos \theta_{yY} + f_3 \cos \theta_{yZ} \\
q_3 &= f_1 \cos \theta_{zX} + f_2 \cos \theta_{zY} + f_3 \cos \theta_{zZ}
\end{aligned}
$$



In matrix form,

$$
\mathbf{q} = \mathbf{R} \mathbf{f}
$$


$$
\begin{bmatrix}
q_1 \\
q_2 \\
q_3
\end{bmatrix}
=
\begin{bmatrix}
r_{xX} & r_{xY} & r_{xZ} \\
r_{yX} & r_{yY} & r_{yZ} \\
r_{zX} & r_{zY} & r_{zZ}
\end{bmatrix}
\begin{bmatrix}
f_1 \\
f_2 \\
f_3
\end{bmatrix}
$$

where $\mathbf{R}$ is the $3 \times 3$ rotation matrix, and  the coefficients $r_{ij}$ are the **direction cosines** relating the local and global axes.

$$
r_{ij} = \cos \theta_{ij},
\qquad i = x, y, z,
\qquad j = X, Y, Z
$$

### 

### Resolving Global Moments Into Local Components

The local moment components at the start node, $b$, are obtained in the same way, by projecting the global moment components onto the local element axes.

$$
\begin{bmatrix}
q_4 \\
q_5 \\
q_6
\end{bmatrix}
=
\begin{bmatrix}
r_{xX} & r_{xY} & r_{xZ} \\
r_{yX} & r_{yY} & r_{yZ} \\
r_{zX} & r_{zY} & r_{zZ}
\end{bmatrix}
\begin{bmatrix}
f_4 \\
f_5 \\
f_6
\end{bmatrix}
$$

So the same direction-cosine matrix is used to transform both force components and moment components from global to local coordinates.

This is the same for DOFs 7 through 12.


### Transformation Matrix for a Space Frame Element

Collecting the force and moment transformations for both element ends leads to the full member transformation relation between the **local** and **global** end-force vectors:

$$
\mathbf{q} = \mathbf{T}\mathbf{f}
$$

The transformation matrix is built from copies of the same $3 \times 3$ rotation matrix $\mathbf{R}$.

$$
\mathbf{T} =
\begin{bmatrix}
\mathbf{R} & \mathbf{0} & \mathbf{0} & \mathbf{0} \\
\mathbf{0} & \mathbf{R} & \mathbf{0} & \mathbf{0} \\
\mathbf{0} & \mathbf{0} & \mathbf{R} & \mathbf{0} \\
\mathbf{0} & \mathbf{0} & \mathbf{0} & \mathbf{R}
\end{bmatrix}
$$

So the same $3 \times 3$ rotation block is used for:

- forces at start node, $b$
- moments at start node, $b$
- forces at end node, $e$
- moments at end node, $e$

### Summary of Transformations

Once the member transformation matrix $\mathbf{T}$ is defined, we can move consistently between the **local** and **global** coordinate systems.


Global nodal displacements are transformed to local element displacements by $\mathbf{v} = \mathbf{T}\mathbf{u}$

Global element end forces are transformed to local element end forces by $\mathbf{q} = \mathbf{T}\mathbf{f}$

Global FEFs are transformed to local FEFs by $\mathbf{q}^F = \mathbf{T} \mathbf{f}^F$

Local element stiffness matrices are transformed to global element stiffness matrices by $\mathbf{K} = \mathbf{T}^T \mathbf{k}\mathbf{T}$

Local nodal displacements are transformed to global element displacements by $\mathbf{u} = \mathbf{T}^T\mathbf{v}$

Local element end forces are transformed to global element end forces by $\mathbf{f} = \mathbf{T}^T \mathbf{q}$

Local FEFs are transformed to global fixed-end-force vectors by $\mathbf{f}^F = \mathbf{T}^T \mathbf{q}^F$



### Calculating the $3 \times 3$ Rotation Matrix $\mathbf{R}$

Review **Lecture 5.1** for the basic construction used for **3D truss elements**.

That same construction gives a valid $3 \times 3$ rotation matrix for a general 3D member: the member end coordinates define the local $x$-axis, and a local frame can then be constructed from it.

For a **3D frame element**, however, this is generally **not enough**. The member geometry defines local $x$, but it does not uniquely define the local $y$- and $z$-axes, since the cross section may still be rotated about the member axis.

Thus, the Lecture 5.1 construction is directly sufficient only when this additional rotation is irrelevant, such as for **3D trusses** or members with **axisymmetric cross-sections** (i.e., circular)


### Additional Orientation Information

For a **3D frame element**, the local coordinate system requires:

- the member axis
- the **rotation of the cross section** about that axis

In the **Lecture 5.1** construction, the cross-product procedure defines a **reference local frame**. A **roll angle** then describes the additional rotation about local $x$ needed to obtain the actual cross-sectional orientation.

<div style="text-align:center;">
  <img src="assets/L1_RollAngle.png" style="width:20%;">
</div>

Without this extra information, the local $y$- and $z$-axes are not uniquely defined.


### Complete $3 \times 3$ Rotation Matrix $\mathbf{R}$ for Frames

This form expresses the full rotation matrix in terms of the member’s local $x$-axis direction cosines and a roll angle $\Psi$ about that axis.

$$
\begin{bmatrix}
r_{xX} & r_{xY} & r_{xZ} \\[8pt]
\dfrac{-r_{xX} r_{xY} \cos \Psi - r_{xZ} \sin \Psi}{\sqrt{r_{xX}^2 + r_{xZ}^2}} &
\sqrt{r_{xX}^2 + r_{xZ}^2}\cos \Psi &
\dfrac{-r_{xY} r_{xZ} \cos \Psi + r_{xX} \sin \Psi}{\sqrt{r_{xX}^2 + r_{xZ}^2}} \\[12pt]
\dfrac{r_{xX} r_{xY} \sin \Psi - r_{xZ} \cos \Psi}{\sqrt{r_{xX}^2 + r_{xZ}^2}} &
-\sqrt{r_{xX}^2 + r_{xZ}^2}\sin \Psi &
\dfrac{r_{xY} r_{xZ} \sin \Psi + r_{xX} \cos \Psi}{\sqrt{r_{xX}^2 + r_{xZ}^2}}
\end{bmatrix}
$$

### In This Course

A full 3D frame formulation requires additional information to define the cross-sectional rotation about the member axis. 

Implementing this in code is beyond the scope of this course.

Therefore, we will use the same basic $3 \times 3$ rotation-matrix construction introduced in **Lecture 5.1** for **3D truss elements**.

For more details on the full derivation of $\textbf{R}$ for a 3D frame element, see **Kassimali, 8.3**.


## Part 6 - Member Releases

### Types of Releases

In general, a 3D frame element may include many possible combinations of end moment releases. The modified equilibrium equations can be derived using the same basic procedure introduced earlier for **planar frame elements** in **Lecture 8**.

Because the number of possible release combinations is much larger in 3D, we will restrict attention to the simplest cases, chosen to mirror the 2D frame formulation:

- **MT = 0:** no hinges (rigid at both ends, normal 3D frame element)
- **MT = 1:** hinge at start node (**b**), all 3 moments released (ball and socket connection)
- **MT = 2:** hinge at end node (**e**), all 3 moments released (ball and socket connection)
- **MT = 3:** hinges at **both ends**, all 6 moments released (reduces to 3D truss element)
