## Part 1 — Beam Element Stiffness Relations

###  Beam Element Response

The **member stiffness relations** express the forces at the ends of a beam element  
(including shear forces and bending moments) as functions of the **end displacements**  
(including transverse displacements and rotations).

When a beam element is subjected to external loading:

- The member **deforms** (bending)
- **Internal shear forces** develop at the ends  
- **Bending moments** are induced at the ends  

These internal forces are fully determined by the **displacements and rotations at the element ends**.

### Generic Displacement for a 2D Beam Element

We define all quantities in the **local coordinate system**, with origin at the left end $b$, and ending at node $e$.

**Sign conventions:**

- $u > 0$ → upward (local $y$ direction)  
- $\theta > 0$ → counterclockwise  
- Forces follow the same order: $[V_b,\ M_b,\ V_e,\ M_e]^T$

### Local displacement and force vectors

Local displacement vector (including rotations):
$$
\boldsymbol{u}=
\begin{Bmatrix}
u_b\\
\theta_b\\
u_e\\
\theta_e
\end{Bmatrix}
=
\begin{Bmatrix}
u_1\\
u_2\\
u_3\\
u_4
\end{Bmatrix}
$$

Local nodal force vector (including moments):
$$
\boldsymbol{Q} =
\begin{Bmatrix}
V_b\\
M_b\\
V_e\\
Q_4
\end{Bmatrix}=
\begin{Bmatrix}
Q_1\\
Q_2\\
Q_3\\
Q_4
\end{Bmatrix}
$$

We seek:
$$\boldsymbol{Q} = \boldsymbol{k}\,\boldsymbol{u}$$


### Four equations (one per DOF)

recall, this form is similar to the definition for a truss element.

$$
Q_1 = k_{11}u_1 + k_{12}u_2 + k_{13}u_3 + k_{14}u_4
$$
$$
Q_2 = k_{21}u_1 + k_{22}u_2 + k_{23}u_3 + k_{24}u_4
$$
$$
Q_3 = k_{31}u_1 + k_{32}u_2 + k_{33}u_3 + k_{34}u_4
$$
$$
Q_4 = k_{41}u_1 + k_{42}u_1 + k_{43}u_3 + k_{44}u_4
$$

Each equation expresses **force equilibrium at a single local degree of freedom**.

For a linear elastic element, the force at any DOF is a **linear combination of all DOF displacements**:
- displacing one DOF can induce forces at *all* DOFs
- the proportionality constants are the stiffness coefficients $k_{ij}$

### Same equations in matrix form

This form is similar to a truss element.

$$
\begin{Bmatrix}
Q_1\\
Q_2\\
Q_3\\
Q_4
\end{Bmatrix}
=
\begin{bmatrix}
k_{11} & k_{12} & k_{13} & k_{14}\\
k_{21} & k_{22} & k_{23} & k_{24}\\
k_{31} & k_{32} & k_{33} & k_{34}\\
k_{41} & k_{42} & k_{43} & k_{44}
\end{bmatrix}
\begin{Bmatrix}
u_1\\
u_2\\
u_3\\
u_4
\end{Bmatrix}
$$

## Part 2 — The Beam Element Stiffness Matrix

### Unit Displacement Method Derivation

Recall:

$$
k_{ij} = \text{force at DOF } i \text{ due to a unit displacement at DOF } j, \text{with all other DOFs held fixed.}
$$

Each column of $k$ is built by:

- Impose a **unit displacement** at one DOF
- Hold all the other DOF fixed
- Record the resulting nodal force pattern

We will do this for DOFs 1–4.

### Governing Equation (Beam Bending)

From mechanics of materials, the beam bending equation is:

\begin{equation}
\frac{d^2 u}{dx^2} = \frac{M(x)}{EI}
\tag{1}
\end{equation}

- $u(x)$ = transverse displacement  
- $M(x)$ = bending moment  
- $EI$ = flexural rigidity  

### Step 1 — Express Internal Moment as a Function of $x$

<div style="display:flex; gap:1.5rem; align-items:center;">
  <!-- Figure -->
  <div style="flex:0.5;">
    <figure style="margin:0; text-align:center;">
      <img src="assets/L2_FBD.png" style="width:100%; height:auto;">
    </figure>
  </div>

  <!-- Text -->
  <div style="flex:0.5; font-size:0.95em;">

Cut the beam at a distance $x$ from node $b$.

Using equilibrium:

\begin{equation}
M(x) = -k_{21} + k_{11} x
\tag{2}
\end{equation}

- $k_{21}$ → moment at node $b$  
- $k_{11}$ → shear at node $b$  

  </div>
</div>

### Step 2 — Substitute into Governing Equation

Substitute the expression for the **bending moment** from Eq. (2) into the moment term of Eq. (1):

\begin{equation}
\frac{d^2 u}{dx^2} = \frac{M(x)}{EI}
\;\;\xrightarrow{\text{Eq. (2)}}\;\;
\frac{d^2 u}{dx^2} = \frac{1}{EI} \left(-k_{21} + k_{11} x \right)
\tag{3}
\end{equation}


### Step 3 — Integrate to Obtain Rotation $\theta(x)$

Start from Eq. (3):

\begin{equation}
\frac{d^2 u}{dx^2} = \frac{1}{EI}\left(-k_{21} + k_{11}x\right)
\tag{3}
\end{equation}

Integrate **both sides** with respect to $x$:

\begin{equation}
\int \frac{d^2 u}{dx^2}\,dx
=
\int \frac{1}{EI}\left(-k_{21} + k_{11}x\right)\,dx
\tag{4a}
\end{equation}

Left-hand side:

\begin{equation}
\int \frac{d^2 u}{dx^2}\,dx = \frac{du}{dx} = \theta(x)
\tag{4b}
\end{equation}



Right-hand side (integrate term-by-term, and pull out constants $1/(EI)$):

\begin{equation}
\int \frac{1}{EI}\left(-k_{21} + k_{11}x\right)\,dx
=
\frac{1}{EI}\left(\int -k_{21}\,dx + \int k_{11}x\,dx\right)
\tag{4c}
\end{equation}

Compute each integral:

\begin{equation}
\int -k_{21}\,dx = -k_{21}x
\qquad
\int k_{11}x\,dx = k_{11}\frac{x^2}{2}
\tag{4d}
\end{equation}

Include the integration constant:

\begin{equation}
\theta(x)=\frac{du}{dx}
=
\frac{1}{EI}\left(-k_{21}x+\frac{k_{11}}{2}x^2\right)+C_1
\tag{4}
\end{equation}


### Step 4 — Integrate to Obtain Displacement $u(x)$

Start from Eq. (4):

\begin{equation}
\frac{du}{dx} =
\frac{1}{EI}\left(-k_{21}x+\frac{k_{11}}{2}x^2\right)+C_1
\tag{4}
\end{equation}

Integrate **both sides** with respect to $x$:

\begin{equation}
\int \frac{du}{dx}\,dx
=
\int \left[
\frac{1}{EI}\left(-k_{21}x+\frac{k_{11}}{2}x^2\right)+C_1
\right] dx
\tag{5a}
\end{equation}

Left-hand side:

\begin{equation}
\int \frac{du}{dx}\,dx = u(x)
\tag{5b}
\end{equation}



Right-hand side (integrate term-by-term):

\begin{equation}
u(x)=
\frac{1}{EI}\left(\int -k_{21}x\,dx + \int \frac{k_{11}}{2}x^2\,dx\right)
+ \int C_1\,dx
\tag{5c}
\end{equation}


Compute each integral:

\begin{equation}
\int -k_{21}x\,dx = -\frac{k_{21}}{2}x^2
\qquad
\int \frac{k_{11}}{2}x^2\,dx = \frac{k_{11}}{6}x^3
\qquad
\int C_1\,dx = C_1 x
\tag{5d}
\end{equation}

Include the second integration constant:

\begin{equation}
u(x)=
\frac{1}{EI}\left(-\frac{k_{21}}{2}x^2+\frac{k_{11}}{6}x^3\right)
+ C_1 x + C_2
\tag{5}
\end{equation}


### Step 5 — Apply Boundary Conditions

We enforce **unit displacement at node $b$** (0 elsewhere):

At $x = 0$:
\begin{align*}
\theta(0) &= 0 \\
u(0) &= 1
\end{align*}

At $x = L$:
\begin{align*}
\theta(L) &= 0 \\
u(L) &= 0
\end{align*}

### Step 6 — Solve for Constants

Substitute $x=0$ into Eqs. (4)–(5):

- From $\theta(0) = 0$: all terms with $x$ vanish $\Rightarrow C_1 = 0$  
- From $u(0) = 1$: all terms with $x$ vanish $\Rightarrow C_2 = 1$  


Updated Eqs. (4)–(5):

\begin{equation}
\theta(x) = \frac{1}{EI} \left(-k_{21} x + \frac{k_{11}}{2} x^2 \right)
\tag{6}
\end{equation}

\begin{equation}
u(x) =
\frac{1}{EI} \left(-\frac{k_{21}}{2} x^2 + \frac{k_{11}}{6} x^3 \right) + 1
\tag{7}
\end{equation}


### Step 7 — Apply Boundary Conditions at $x = L$


Substitute $\theta(L) = 0$ into Eq. (6):

\begin{equation*}
0 = \frac{1}{EI} \left(-k_{21} L + \frac{k_{11}}{2} L^2 \right)
\end{equation*}

Multiply both sides by $EI$ (nonzero), so it cancels:

\begin{equation*}
0 = -k_{21} L + \frac{k_{11}}{2} L^2
\end{equation*}

Rearrange to solve for $k_{21}$:

\begin{equation*}
k_{21} L = \frac{k_{11}}{2} L^2
\end{equation*}

Divide by $L$:

\begin{equation*}
k_{21} = \frac{k_{11}}{2} L
\tag{9}
\end{equation*}

### Step 8 — Apply Final Condition

Substitute $u(L)=0$ into Eq. (7):

\begin{equation*}
0 =
\frac{1}{EI}\left(-\frac{k_{21}}{2}L^2+\frac{k_{11}}{6}L^3\right)+1
\end{equation*}

Rearrange:

\begin{equation*}
-EI =
-\frac{k_{21}}{2}L^2+\frac{k_{11}}{6}L^3
\end{equation*}

Now substitute Eq. (9), $k_{21}=\dfrac{k_{11}L}{2}$:

\begin{equation*}
-EI =
-\frac{1}{2}\left(\frac{k_{11}L}{2}\right)L^2+\frac{k_{11}}{6}L^3
\end{equation*}

Solve for $k_{11}$:

\begin{equation*}
\boxed{
k_{11} = \frac{12EI}{L^3}
}
\end{equation*}



### Step 9 — Final Coefficient

Back-substitute $k_{11}$ into Eq. (9):

\begin{equation*}
\boxed{
k_{21} = \frac{6EI}{L^2}
}
\end{equation*}

### Step 10 — Solve Remaining Stiffness Terms $k_{31}$ and $k_{41}$

Use equilibrium of the **beam element free-body diagram** (with the previously found terms
$k_{11}=\dfrac{12EI}{L^3}$ and $k_{21}=\dfrac{6EI}{L^2}$).

<div style="display:flex; gap:1.5rem; align-items:center;">

  <!-- Text -->
  <div style="flex:0.55; font-size:0.95em; line-height:1.25;">

**(1) Vertical force equilibrium**

\begin{equation*}
\sum F_y = 0:
\qquad
\frac{12EI}{L^3} + k_{31} = 0
\end{equation*}

\begin{equation*}
\boxed{
k_{31} = -\frac{12EI}{L^3}
}
\end{equation*}

**(2) Moment equilibrium about end $b$**

\begin{equation*}
\sum M_b = 0:
\qquad
\frac{6EI}{L^2} - \left(\frac{12EI}{L^3}\right)L + k_{41} = 0
\end{equation*}

\begin{equation*}
\boxed{
k_{41} = \frac{6EI}{L^2}
}
\end{equation*}

### Step 11 (Optional) — Displacement Shape Function

Substitute the solved coefficients $k_{11}$ and $k_{21}$ into Eq. (7):

\begin{equation*}
u(x) =
\frac{1}{EI} \left(-\frac{k_{21}}{2} x^2 + \frac{k_{11}}{6} x^3 \right) + 1
\;\;\xrightarrow{\;k_{11},\,k_{21}\;}
\;\;
u(x) = 1 - 3\left(\frac{x}{L}\right)^2 + 2\left(\frac{x}{L}\right)^3
\end{equation*}
