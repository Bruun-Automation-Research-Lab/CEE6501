## What is a matrix?

A matrix is a compact way to represent a **linear relationship** between vectors.

$$
\mathbf{y} = \mathbf{A}\mathbf{x}
$$

- $\mathbf{x}\in\mathbb{R}^n$: input (e.g., displacements)
- $\mathbf{y}\in\mathbb{R}^m$: output (e.g., forces)
- $\mathbf{A}\in\mathbb{R}^{m\times n}$: mapping (e.g., stiffness)

**In structural analysis:**
- $\mathbf{K}\mathbf{u} = \mathbf{f}$ (global stiffness)
- $\mathbf{B}\mathbf{u} = \boldsymbol{\varepsilon}$ (strain–displacement)
- $\mathbf{A}^T\boldsymbol{\sigma} = \mathbf{f}$ (equilibrium / virtual work forms)
