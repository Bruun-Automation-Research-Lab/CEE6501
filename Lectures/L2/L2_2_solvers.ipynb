## Learning Objectives (Section 2)

By the end of this section, you will:
- Understand what it means to **solve a system of linear equations**
- Distinguish between **direct solution methods** and matrix factorizations
- Recognize common factorizations (LU, Cholesky, LDLᵀ) and their assumptions
- Appreciate how **matrix properties** affect solvability and numerical behavior
