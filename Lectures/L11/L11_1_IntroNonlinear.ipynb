## Part 1 — What Is Nonlinearity?

> We begin by stepping beyond the assumptions of ordinary linear structural analysis.

### Review: What Linear Analysis Assumes

In the usual **linear** structural analysis, we rely on two key assumptions:

- **Geometric linearity**: deformations are small enough that equilibrium can be written on the **undeformed geometry**
- **Material linearity**: stress is proportional to strain, so the material remains **linearly elastic**

Under these assumptions:

- stiffness does not depend on the current deformed state
- load and displacement are proportional
- the principle of superposition holds

This is often adequate for many structures under ordinary **service-load** conditions.

### Why Nonlinear Analysis Is Needed

As loads increase, or as deformations become significant, the assumptions of linear analysis begin to break down.

A structure may become nonlinear because:

- its **geometry changes enough** during deformation to affect equilibrium and internal force
- its **material behavior changes**, for example through yielding, cracking, or inelastic response

Then the structural response is no longer a straight-line relation between load and displacement.



In nonlinear analysis:

- stiffness may change during loading
- equilibrium is written on the current or updated configuration
- the response may involve instability, bifurcation, limit points, or post-critical behavior


### Why This Matters in Structural Engineering

Linear analysis is often very useful, but it has inherent limitations.

It may not capture:

- loss of stiffness caused by large deformations
- second-order effects such as **$P$-$\Delta$** and **$P$-$\delta$**
- instability phenomena such as buckling
- inelastic material behavior and strength degradation

So nonlinear analysis becomes important when the goal is not only to compute forces and displacements under service conditions, but also to understand:

- ultimate strength
- stability
- large-deformation behavior
- failure mechanisms


## Part 2 — Types of Nonlinearity

### Two Main Types

We can classify most structural nonlinearity into two broad categories:

**Material nonlinearity:** The constitutive law is nonlinear, so the relationship between stress and strain is no longer linear.

**Geometric nonlinearity:** The deformation is large enough that the changing geometry affects equilibrium, compatibility, and internal force.

A real structure may exhibit:

- only material nonlinearity
- only geometric nonlinearity
- or both simultaneously


### Geometric Nonlinearity

Geometric nonlinearity arises when deformations become large enough that the current geometry cannot be ignored.

Typical examples include:

- $P$-$\Delta$ effect: gravity load acting through a lateral story displacement
- $P$-$\delta$ effect: axial force interacting with member curvature
- cables, membranes, or very flexible trusses whose stiffness depends strongly on shape (this is important for us)
- structures undergoing large displacement before reaching material failure

Conceptually:

- equilibrium must be written on the **deformed** configuration
- internal force may depend on both displacement and current axial state
- stiffness becomes configuration-dependent

### Material Nonlinearity

Material nonlinearity arises when the stress--strain law is not linear elastic.

Typical examples include:

- yielding of steel
- cracking of reinforced concrete
- creep or time-dependent deformation
- nonlinear interaction of axial force, bending, shear, and torsion

Conceptually:

- the material stiffness changes as loading proceeds
- unloading and reloading may follow different paths
- the load-displacement curve can become history-dependent

### Combined Nonlinear Effects

In many realistic structural problems, the two types of nonlinearity occur together.

Examples include:

- a steel frame with both **yielding** and **$P$-$\Delta$ effects**
- reinforced concrete members with **cracking** and **large deflections**
- frames where connection deformation, panel-zone deformation, and stability effects all interact

This is important conceptually, but in a first introduction it is common to isolate one source of nonlinearity at a time.

## Part 3 — Levels of Analysis

### Why We Talk About Levels of Analysis

It is rarely practical to model every source of nonlinearity in full detail.

So structural analysis is often organized into **levels of analysis**, where each level captures more of the true behavior than the one before it.

These levels differ in:

- what assumptions are retained
- what sources of nonlinearity are included
- what kinds of failure or instability they can detect
- the computational effort required


### First-Order Elastic Analysis

This is the standard **linear** analysis we have used so far.

Assumptions:

- equilibrium written on the **undeformed** structure
- material remains **linearly elastic**
- stiffness remains constant

Strengths:

- simple and efficient
- often adequate for service-load response

Limitations:

- does **not** capture geometric nonlinear effects
- does **not** capture material nonlinear effects
- does **not** reveal instability behavior reliably

### Second-Order Elastic Analysis

This level includes the effects of deformation on equilibrium while still assuming elastic material behavior.

So it can represent:

- finite displacement effects
- $P$-$\Delta$ effects
- $P$-$\delta$ effects
- some aspects of stability and load amplification

But it still does **not** represent yielding, cracking, or other material nonlinearity.

This is the level most closely connected to the geometric nonlinear ideas we will begin developing in this part of the course.

### First-Order Inelastic and Second-Order Inelastic Analysis

More advanced levels of analysis include material nonlinearity, with or without geometric nonlinearity.

First-order inelastic analysis:

- includes material nonlinearity
- keeps equilibrium on the undeformed geometry
- can represent plastic response, but not geometric nonlinear instability effects

Second-order inelastic analysis:

- includes both **material** and **geometric** nonlinearity
- is the most general of these common levels
- can represent strength loss, stability limits, and strongly nonlinear response

## Summary of the Levels of Analysis

| Level of analysis | Geometric nonlinearity? | Material nonlinearity? | Typical use |
|---|---:|---:|---|
| First-order elastic | No | No | Basic service-load response |
| Second-order elastic | Yes | No | Large-displacement elastic response, stability effects |
| First-order inelastic | No | Yes | Inelastic material response without geometric effects |
| Second-order inelastic | Yes | Yes | Most general response including strength and stability |

## Part 4 — Scope for This Course

### What We Will Focus On

In CEE6501, we will **not** attempt a general treatment of all nonlinear structural analysis.

Instead, we will focus on:

- geometric nonlinearity
- axial elements only
- bar and joint frameworks (trusses, space frames, cable nets)
- initially in 2D plane trusses

So the goal is not a full nonlinear analysis of general frames or continues surfaces.

The goal is to understand the **basic concepts of geometrically nonlinear analysis** for axial member structures whose stiffness and equilibrium depend strongly on their deformed shape.

### Why This Narrower Scope Makes Sense



This narrower focus is well suited to the themes of this course.

For flexible structural systems with large deformations:

- geometry can dominate the response
- equilibrium depends on the current shape
- finding the correct deformed configuration is itself part of the problem
