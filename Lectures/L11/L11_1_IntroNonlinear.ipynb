## Part 1 — What Is Nonlinearity?

> We begin by stepping beyond the assumptions of ordinary linear structural analysis.

### Review: What Linear Analysis Assumes

In the usual **linear** structural analysis, we rely on two key assumptions:

- **Geometric linearity**: deformations are small enough that equilibrium can be written on the **undeformed geometry**
- **Material linearity**: stress is proportional to strain, so the material remains **linearly elastic**



In its simplest form, linear structural analysis gives a relation such as

$$
\mathbf{K}\mathbf{u} = \mathbf{f}
$$

where:

- $\mathbf{K}$ is constant
- response is proportional to load
- superposition holds

This is often adequate for many structures under ordinary **service-load** conditions.

### A Simple Mathematical View

In a **linear** structural problem, the applied load and the response are proportional.

For a single degree of freedom spring:

$$
f = ku
$$

where:

- $f$ = applied force
- $u$ = displacement
- $k$ = stiffness

If $k$ is constant, then doubling the load doubles the displacement:

$$
2f = k \times 2u
$$

So the force-displacement curve is a **straight line**.


In a **nonlinear** problem, that proportional relationship no longer holds.  

The response may instead look like.

$$
f = f(u)
$$

which means the force is some nonlinear function of displacement.

For example, the slope of a menber may change as the structure deforms, so the stiffness is no longer constant:

$$
k_{\text{tangent}} = \frac{df}{du}
$$



At the structural level, we usually write the problem in terms of **equilibrium** between external and internal forces:

$$
\mathbf{R}(\mathbf{u}) = \mathbf{f}_{\mathrm{ext}} - \mathbf{f}_{\mathrm{int}}(\mathbf{u}) = \mathbf{0}
$$

Here:

- $\mathbf{f}_{\mathrm{ext}}$ is the applied load vector
- $\mathbf{f}_{\mathrm{int}}(\mathbf{u})$ is the internal resisting force vector
- $\mathbf{R}(\mathbf{u})$ is the **residual**, or out-of-balance force

So in a nonlinear analysis, the goal is to find the displacement state $\mathbf{u}$ for which the structure is in equilibrium.

This is harder than the linear case because:

- internal force depends on the current displacement
- stiffness may change as the structure deforms
- the solution usually must be found **iteratively**

### Why Nonlinear Analysis Is Needed

As loads increase, or as deformations become significant, the assumptions of linear analysis begin to break down.

A structure may become nonlinear because:

- its **geometry changes enough** during deformation to affect equilibrium and internal force
- its **material behavior changes**, for example through yielding, cracking, or inelastic response

Then the structural response is no longer a straight-line relation between load and displacement.

<b>Key idea:</b> in a nonlinear problem, the structure you are solving for also changes while you solve it.

### What Changes in Nonlinear Analysis?

In nonlinear analysis:

- stiffness may change during loading
- equilibrium may be written on the **current** or **updated** configuration
- the load-displacement path may curve, soften, or stiffen
- the response may involve instability, bifurcation, limit points, or post-critical behavior

So instead of solving once with a fixed stiffness matrix, we usually solve **iteratively**.

### Why This Matters in Structural Engineering

Linear analysis is often very useful, but it has inherent limitations.

It may not capture:

- loss of stiffness caused by large deformations
- second-order effects such as **$P$-$\Delta$** and **$P$-$\delta$**
- instability phenomena such as buckling
- inelastic material behavior and strength degradation

So nonlinear analysis becomes important when the goal is not only to compute forces and displacements under service conditions, but also to understand:

- ultimate strength
- stability
- large-deformation behavior
- failure mechanisms


## Part 2 — Types of Nonlinearity

### Material Nonlinearity

One major source of structural nonlinearity is **material nonlinearity**.

The constitutive law is nonlinear, so the relationship between stress and strain is no longer a straight line.

For a linear elastic material, we write

$$
\sigma = E\varepsilon
$$

where the slope $E$ is constant.

For a nonlinear material, the stress is some more general function of strain:

$$
\sigma = \sigma(\varepsilon)
$$

So the material stiffness is no longer constant. Instead, the slope changes as the material deforms:

$$
E_t = \frac{d\sigma}{d\varepsilon}
$$

### Examples of Material Nonlinearity

Material nonlinearity arises when the stress--strain law is not linear elastic.

Typical examples include:

- yielding of steel
- cracking of reinforced concrete
- creep or time-dependent deformation
- nonlinear interaction of axial force, bending, shear, and torsion

Conceptually:

- the material stiffness changes as loading proceeds
- unloading and reloading may follow different paths
- the load-displacement curve can become history-dependent

<div style="margin-top:0.5em;">
For a linear elastic material, the slope is constant. For a nonlinear material, the tangent stiffness changes along the response path.
</div>

### Geometric Nonlinearity

The second major source is **geometric nonlinearity**.

The deformation is large enough that the changing geometry affects equilibrium, compatibility, and internal force.

In a linear analysis, we treat the structure as if its geometry does not change significantly during loading, so the stiffness is taken as fixed.

In a geometrically nonlinear analysis, the current deformed shape affects the response, so the stiffness depends on displacement:

$$
\mathbf{K} = \mathbf{K}(\mathbf{u})
$$

Equivalently, the internal resisting force depends on the current configuration:

$$
\mathbf{f}_{\mathrm{int}} = \mathbf{f}_{\mathrm{int}}(\mathbf{u})
$$

### Examples of Geometric Nonlinearity

Geometric nonlinearity arises when deformations become large enough that the current geometry cannot be ignored.

Typical examples include:

- $P$-$\Delta$ effect: gravity load acting through a lateral story displacement
- $P$-$\delta$ effect: axial force interacting with member curvature
- cables, membranes, or very flexible trusses whose stiffness depends strongly on shape
- structures undergoing large displacement before reaching material failure

Conceptually:

- equilibrium must be written on the **deformed** configuration
- internal force depends on both displacement and current axial state
- stiffness becomes configuration-dependent


### Combined Nonlinear Effects

In many realistic structural problems, the two types of nonlinearity occur together.

Examples include:

- a steel frame with both **yielding** and **$P$-$\Delta$ effects**
- reinforced concrete members with **cracking** and **large deflections**
- frames where connection deformation, panel-zone deformation, and stability effects all interact

A useful way to think about it is:

- **material nonlinearity** changes the constitutive law
- **geometric nonlinearity** changes the equilibrium problem

In a first introduction, it is common to isolate one source of nonlinearity at a time.

## Part 3 — Levels of Analysis

### Why We Talk About Levels of Analysis

It is rarely practical to model every source of nonlinearity in full detail.

So structural analysis is often organized into **levels of analysis**, where each level captures more of the true behavior than the one before it.

These levels differ in:

- what assumptions are retained
- what sources of nonlinearity are included
- what kinds of failure or instability they can detect
- the computational effort required

Each higher level is not just "more complicated" — it is answering a different question about the structure.

### Second-Order Elastic Analysis

This level includes the effects of deformation on equilibrium while still assuming elastic material behavior.

So it can represent:

- finite displacement effects
- $P$-$\Delta$ effects
- $P$-$\delta$ effects
- some aspects of stability and load amplification

But it still does **not** represent yielding, cracking, or other material nonlinearity.

Symbolically, the problem begins to look like

$$
\mathbf{K}(\mathbf{u}) \times \mathbf{u} = \mathbf{f}
$$

or, more carefully, as a sequence of updated equilibrium states.

Several possible modes of nonlinear elastic behavior are shown.

- **bifurcation** of the load path into an alternative branch
- gradual softening that culminates in an **elastic instability** at a limit point
- stiffness changes that may lead to either softening or stiffening behavior

## Part 5 — Basic Concept of Geometrically Nonlinear Analysis

### The Example Structure

To illustrate the basic concept of **geometrically nonlinear analysis**, we consider a simple symmetric **two-member plane truss** made of a linearly elastic material.

The structure is loaded by a vertical load $P$ at the top joint, so the motion is described by a single degree of freedom:

$$
\delta = \text{vertical displacement of joint 2}
$$

<div style="text-align:center;">
  <img src="assets/L1_2Bar1.png" style="width:70%;">
</div>

Because the structure is symmetric:

- both members have the same undeformed length $L$
- both members have the same area $A$
- both members have the same angle $\theta$ in the undeformed configuration
- both members carry the same axial force magnitude $Q$

This very simple system is enough to show the essential difference between **linear** and **geometrically nonlinear** analysis.


### Linear Analysis: Small-Displacement Kinematics

In **linear analysis**, the displacement $\delta$ is assumed to be small enough that the member axial deformation can be approximated from the **undeformed** geometry.

For each member,

$$
u \approx \delta \sin\theta
\tag{1}
$$

<div style="text-align:center;">
  <img src="assets/L1_2Bar2.png" style="width:50%;">
</div>


so the axial strain is

$$
\varepsilon = \frac{u}{L} \approx \left(\frac{\sin\theta}{L}\right)\delta
\tag{2}
$$

This is a **linear** relation between strain and joint displacement.

The key approximation is that the member direction is taken from the **original configuration**, not the deformed one ($L = L_0$)


### Linear Analysis: Equilibrium on the Undeformed Geometry

In linear analysis, the equilibrium equations are also written on the **undeformed** geometry.

At the loaded top joint, vertical equilibrium gives

$$
Q \approx \frac{P}{2\sin\theta}
\tag{3}
$$

<div style="text-align:center;">
  <img src="assets/L1_2Bar3.png" style="width:50%;">
</div>


and therefore the axial stress is

$$
\sigma = \frac{Q}{A} \approx \frac{P}{2A\sin\theta}
\tag{4}
$$

For a linearly elastic material,

$$
\sigma = E\varepsilon
\tag{5}
$$

Combining the linear expressions for stress, Eq. (4), and strain, Eq. (2), gives the linear load-displacement relation:

$$
P \approx \underbrace{\left(\frac{2EA\sin^2\theta}{L}\right)}_{\text{stiffness term}} \delta
\tag{6}
$$

So in the linear model, the structure behaves like a spring with constant stiffness:

$$
k = \frac{2EA\sin^2\theta}{L}
$$


### Geometrically Nonlinear Analysis: Use the Deformed Geometry

We do not assume the displacement is small.

<div style="text-align:center;">
  <img src="assets/L1_2Bar4.png" style="width:40%;">
</div>

$$
\bar{L}
=
\sqrt{(L\cos\theta)^2 + (L\sin\theta - \delta)^2}
=
L\sqrt{1+\left(\frac{\delta}{L}\right)^2-2\left(\frac{\delta}{L}\right)\sin\theta}
\tag{7}
$$

$$
\sin\bar{\theta}
=
\frac{L\sin\theta-\delta}{\bar{L}}
=
\frac{\sin\theta-\left(\frac{\delta}{L}\right)}
{\sqrt{1+\left(\frac{\delta}{L}\right)^2-2\left(\frac{\delta}{L}\right)\sin\theta}}
\tag{8}
$$

### Exact Axial Deformation and Strain

The **first key difference** from linear analysis is in the **kinematics**. The member deformation is computed from the **actual deformed length**.

The exact axial deformation is therefore

$$
u = \bar{L} - L
$$

Substituting the expression for the deformed member length from Eq. (7) gives

$$
u
=
L\left[
\sqrt{1+\left(\frac{\delta}{L}\right)^2-2\left(\frac{\delta}{L}\right)\sin\theta}
-1
\right]
\tag{9}
$$

and the corresponding axial strain becomes

$$
\varepsilon
=
\frac{u}{L}
=
\sqrt{1+\left(\frac{\delta}{L}\right)^2-2\left(\frac{\delta}{L}\right)\sin\theta}
-1
\tag{10}
$$

So, unlike the linear case, both axial deformation and strain are now **nonlinear functions** of the joint displacement $\delta$.

### Equilibrium Must Also Be Written on the Deformed Geometry

The **second** key difference from linear analysis is in the equilibrium equations since when the current member angle changes, so too does the vertical equilibrium relation.


<div style="display:flex; align-items:center; gap:4%;">

<div style="width:49%;">

vertical equilibrium at the top joint

$$
Q = \frac{P}{2\sin\bar{\theta}}
\tag{11}
$$

</div>

<div style="width:49%; text-align:center;">

<img src="assets/L1_2Bar5.png" style="width:100%;">

</div>

</div>


So unlike the linear case, the internal member force is not related to $P$ through the original angle $\theta$, but through the **current** angle $\bar{\theta}$.



Substituting the expression for the deformed angle from Eq. (8) gives

$$
Q
=
\frac{P\sqrt{1+\left(\frac{\delta}{L}\right)^2-2\left(\frac{\delta}{L}\right)\sin\theta}}
{2\left[\sin\theta-\left(\frac{\delta}{L}\right)\right]}
$$

which yields the expression for member axial stress as

$$
\sigma
=
\frac{Q}{A}
=
\frac{P\sqrt{1+\left(\frac{\delta}{L}\right)^2-2\left(\frac{\delta}{L}\right)\sin\theta}}
{2A\left[\sin\theta-\left(\frac{\delta}{L}\right)\right]}
\tag{12}
$$



Finally, by substituting the expressions for strain, Eq. (10), and stress, Eq. (12), into the linear stress-strain relationship $\sigma = E\varepsilon$, we obtain the nonlinear load-displacement relation for the truss:

$$
P
=
2EA\left[\sin\theta-\left(\frac{\delta}{L}\right)\right]
\left[
\frac{1-\sqrt{1+\left(\frac{\delta}{L}\right)^2-2\left(\frac{\delta}{L}\right)\sin\theta}}
{\sqrt{1+\left(\frac{\delta}{L}\right)^2-2\left(\frac{\delta}{L}\right)\sin\theta}}
\right]
\tag{13}
$$

### Result: A Nonlinear Load-Displacement Relation

After combining:

- the exact deformed geometry
- the exact strain expression
- Hooke's law, $\sigma = E\varepsilon$
- and equilibrium on the deformed configuration

we obtain a nonlinear relation between load $P$ and displacement $\delta$.

Symbolically, the important point is:

$$
P = P(\delta)
$$

not

$$
P = k\delta
$$

with constant stiffness.

So even with a **linearly elastic material**, the structural response becomes nonlinear because the geometry changes.

### Summary — Two Key Differences

**Kinematics**

In linear analysis,

$$
u \approx \delta \sin\theta
$$

In geometrically nonlinear analysis,

$$
u = \bar{L} - L
$$

So the member deformation is based on the **exact deformed geometry**, not a small-displacement approximation.

**Equilibrium**

In linear analysis, equilibrium is written on the **undeformed** geometry.

In geometrically nonlinear analysis, equilibrium is written on the **deformed** geometry.

So both the deformation and the equilibrium equations depend on the **current shape** of the structure.

## Example — Linear vs. Geometrically Nonlinear Analysis

### Example Setup

Consider the shallow two-member truss shown below. 

We will compare:

- the **linear** load-displacement prediction
- the **geometrically nonlinear** load-displacement prediction

for the same structure and material properties.

<div style="text-align:center;">
  <img src="assets/L1_2Bar_Example.png" style="width:60%;">
</div>


```python
import numpy as np
import matplotlib.pyplot as plt

# Two-bar truss properties
E = 70e3            # MPa = N/mm^2
A = 645.2           # mm^2
L = 3000            # mm
theta = np.deg2rad(30)

print(f"E = {E:.0f} MPa")
print(f"A = {A:.1f} mm^2")
print(f"L = {L:.0f} mm")
print(f"theta = {np.rad2deg(theta):.1f} deg")
```

```python
delta = np.arange(0, 4000, 5)  # mm

# Linear model
P_lin = (2 * E * A * np.sin(theta)**2 / L) * delta

# Geometrically nonlinear model from the exact truss geometry
x = delta / L
root = np.sqrt(1 + x**2 - 2*x*np.sin(theta))
P_geo = 2 * E * A * (np.sin(theta) - x) * (1 - root) / root
```

### Equations Used in the Example

The **linear** model is

$$
P_{\mathrm{lin}} = \left(\frac{2EA\sin^2\theta}{L}\right)\delta
\tag{6}
$$

The **geometrically nonlinear** model is based on the exact relation

$$
P
=
2EA\left[\sin\theta-\left(\frac{\delta}{L}\right)\right]
\left[
\frac{1-\sqrt{1+\left(\frac{\delta}{L}\right)^2-2\left(\frac{\delta}{L}\right)\sin\theta}}
{\sqrt{1+\left(\frac{\delta}{L}\right)^2-2\left(\frac{\delta}{L}\right)\sin\theta}}
\right]
\tag{13}
$$

```python
def plot_response(delta, P_lin, P_geo):
    plt.figure(figsize=(8, 5))
    plt.plot(delta, P_lin / 1e6, label='Linear analysis')
    plt.plot(delta, P_geo / 1e6, label='Geometrically nonlinear analysis')

    plt.axhline(0, color='black', linewidth=1)
    plt.axvline(0, color='black', linewidth=1)

    plt.xlabel('Displacement δ (mm)')
    plt.ylabel('Load P (MN)')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plot_response(delta, P_lin, P_geo)
```

### Interpreting the Response Curve

From the plot, we observe:

- the **linear** solution increases indefinitely with constant slope
- the **geometrically nonlinear** solution gradually softens as displacement increases
- the two predictions agree at small displacement, then diverge more and more as deformation grows

This means the linear approximation becomes progressively less accurate as the geometry change becomes significant.



### Critical Point and Tangent Stiffness

The slope of the nonlinear response curve is the **tangent stiffness**.

At its peak the tangent stiffness becomes zero:

$$
\frac{dP}{d\delta} = 0
$$

When it reaches 0, the structure has reached a **critical** or **limit point**.

<div style="text-align:center;">
  <img src="assets/L1_2Bar_Example_critical.png" style="width:70%;">
</div>



For the shallow two-member truss, the corresponding critical displacement is

$$
\delta_{cr}
= L\left[\sin\theta - \cos\theta\sqrt{\frac{1}{(\cos\theta)^{2/3}} - 1}\right]
\tag{14}
$$

Substituting $\delta_{cr}$ into Eq. (13) gives the critical load $P_{cr}$.

This is the point where the structure loses its ability to resist any further increase in load along the same equilibrium path.

```python
delta_cr = L * (np.sin(theta) - np.cos(theta) * np.sqrt(1 / (np.cos(theta)**(2/3)) - 1))

x_cr = delta_cr / L
root_cr = np.sqrt(1 + x_cr**2 - 2 * x_cr * np.sin(theta))
P_cr = 2 * E * A * (np.sin(theta) - x_cr) * (1 - root_cr) / root_cr

print(f"Critical displacement δ_cr = {delta_cr:.2f} mm")
print(f"Critical load P_cr = {P_cr / 1e6:.3f} MN")
```

### Physical Meaning of the Critical Point

The critical point marks the end of stable load increase along the original equilibrium branch.

Before that point:

- additional displacement requires additional load
- the structure still has positive tangent stiffness

At the critical point:

- the tangent stiffness becomes zero
- the response curve reaches a peak
- the structure is on the verge of instability


### Snap-Through Instability

Beyond the critical point, a further increase in load cannot be sustained on the same branch.

Instead, the truss may undergo **snap-through**, moving suddenly to a very different equilibrium configuration.
