## Main Computational Idea: Convergence to Equilibrium

In linear analysis, we typically solve once:

$$
\mathbf{K}\mathbf{u} = \mathbf{f}
$$

In geometrically nonlinear analysis, the stiffness depends on the current state, so the problem is more like:

$$
\mathbf{R}(\mathbf{u}) = \mathbf{0}
$$

where $\mathbf{R}$ is the **residual** or out-of-balance force vector.

So the solution process becomes iterative:

1. guess a configuration
2. evaluate internal force and residual
3. update the configuration
4. repeat until equilibrium is satisfied to tolerance

This iterative viewpoint will be essential for both nonlinear analysis and form finding.

## Part 6 - Roadmap Beyond 2D



Our progression will be:

- start with **2D trusses** to develop the core ideas cleanly
- then extend the intuition toward **3D frameworks**
- then connect those ideas to **form finding** methods

The conceptual bridge is this:

- nonlinear analysis asks whether a given loaded structure is in equilibrium
- form finding asks what **shape** a flexible structure must take in order to be in equilibrium

So understanding geometric nonlinearity is the foundation for both.
