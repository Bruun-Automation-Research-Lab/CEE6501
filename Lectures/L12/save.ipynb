### Meaning of the Symbols in the Unified Equation

We now define the symbols in the Veenendaal and Block unified expression:

$$
\mathbf{x}_{i+1}
=
\mathbf{x}_i
+
\left(\mathbf{C}^{T}\mathbf{K}\mathbf{C}\right)^{-1}
\left(
\mathbf{p}
-
\mathbf{C}^{T}\mathbf{L}^{-1}\mathbf{F}\mathbf{C}_f\,\mathbf{x}_f
-
\mathbf{C}^{T}\mathbf{L}^{-1}\mathbf{F}\mathbf{C}\,\mathbf{x}_i
\right)
$$

This equation represents **one iteration** of a Newton-Raphson nonlinear form-finding procedure written in a common framework.

### Geometry Variables

$$
\mathbf{x}_i
$$

Current vector of **unknown free-node coordinates** at iteration $i$.

$$
\mathbf{x}_{i+1}
$$

Updated vector of **free-node coordinates** after the next iteration.

$$
\mathbf{x}_f
$$

Vector of **known fixed-node coordinates**.

In other words:

- $\mathbf{x}_i$ contains the coordinates we are solving for
- $\mathbf{x}_f$ contains coordinates that are prescribed (fixed coordinates, like at supports)
- the method updates $\mathbf{x}_i$ to obtain $\mathbf{x}_{i+1}$

For a 2D problem, $\mathbf{x}$ typically stacks the unknown nodal $x$- and $y$-coordinates.  
For a 3D problem, it stacks $x$-, $y$-, and $z$-coordinates.

### Connectivity Matrices

$$
\mathbf{C}
$$

The **branch-node matrix** associated with the **free nodes**.

It describes which free nodes are connected to which branches.

Each row corresponds to a branch, and each column corresponds to a free node.

Its entries are typically:

$$
c(j,i)=
\begin{cases}
+1 & \text{if branch } j \text{ starts at node } i \\
-1 & \text{if branch } j \text{ ends at node } i \\
0 & \text{otherwise}
\end{cases}
$$

$$
\mathbf{C}_f
$$

The branch-node matrix associated with the **fixed nodes**.

So together:

- $\mathbf{C}$ connects branches to unknown free nodes
- $\mathbf{C}_f$ connects branches to known fixed nodes

These matrices encode the **topology** of the network.

### Force, Length, and Load Quantities

$$
\mathbf{p}
$$

Vector of **external nodal loads** applied at the free nodes.

$$
\mathbf{L}
$$

A diagonal matrix of **branch lengths**.

Its diagonal entries are the current lengths of the branches:

$$
\mathbf{L} = \operatorname{diag}(l_1,l_2,\dots,l_m)
$$

$$
\mathbf{F}
$$

A diagonal matrix of **branch forces**.

Its diagonal entries are the current axial branch forces:

$$
\mathbf{F} = \operatorname{diag}(f_1,f_2,\dots,f_m)
$$

So:

- $\mathbf{L}^{-1}\mathbf{F}$ gives force divided by length at the branch level
- this is exactly the quantity that becomes the **force density** in the force density method

### Stiffness-Like Quantity in the Unified Framework

$$
\mathbf{K}
$$

A diagonal matrix of **branch stiffness-like quantities** used in the unified iteration.

Its precise meaning depends on the form-finding method:

- in the **force density method**, the governing quantity is the prescribed **force density**
- in **dynamic relaxation**, the internal force state is updated through an iterative equilibrium process
- in the **stiffness method**, $\mathbf{K}$ is related to the branch axial stiffness behavior

So the unified framework keeps the same overall structure, but the meaning of the internal branch quantities changes depending on the method.

### Interpreting the Matrix Products

The individual matrix products have a clear meaning:

$$
\mathbf{C}^{T}\mathbf{K}\mathbf{C}
$$

This is the **free-node system matrix** for the current iteration.

It plays the role of the matrix that must be inverted to compute the coordinate update.

$$
\mathbf{C}^{T}\mathbf{L}^{-1}\mathbf{F}\mathbf{C}_f\,\mathbf{x}_f
$$

This term represents the effect of the **fixed-node coordinates** on the free-node equilibrium.

$$
\mathbf{C}^{T}\mathbf{L}^{-1}\mathbf{F}\mathbf{C}\,\mathbf{x}_i
$$

This term represents the contribution of the **current internal branch forces** associated with the current free-node coordinates.

So the quantity inside parentheses is essentially:

$$
\text{external load contribution}
-
\text{current internal force contribution}
$$

### Main Takeaway

The unified equation shows that all three methods can be viewed as:

- writing equilibrium on a network of branches and nodes
- separating **free** and **fixed** nodal coordinates
- updating the unknown coordinates iteratively

The main difference is not the network description itself.

The main difference is how each method defines and updates:

- branch forces
- branch stiffness-like quantities
- the relation between geometry and internal force
